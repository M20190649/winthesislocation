```python
user_file = "C:/Users/12sha/Documents/thesislocation/Data/user_month.csv"
user_df = pd.read_csv(user_file)
user_df = user_df.drop(['Unnamed: 0'], axis=1)
user_df = user_df.fillna(0)
#user_df.columns = ['User', 'TopMonth', 'DataMonth1', 'DataMonth2', 'DataMonth3', 'DataMonth4', 'DataMonth5']
user_df.head()
```

```python
# count_notrain = 0
# list_notrain = ""
user_tt_df = pd.DataFrame()

for i in range(0, 182):

    user = str(user_df.loc[i, 'User'])
    train_mn = str(user_df.loc[i, 'TopMonth'])
    test_mn = "na"

    # check if next month data is present
    date_s = str(user_df.loc[i, 'TopMonth']) + "01"
    date_d = datetime(year=int(date_s[0:4]), month=int(date_s[4:6]), day=int(date_s[6:8]))
    next_mn = date_d + relativedelta(months=1)
    next_mn = next_mn.strftime('%Y%m%d')
    next_mn = next_mn[:-2]
    next_mn_fl = False
    for j in range(1, 53):
        col_nam = "DataMonth" + str(j)
        if str(int(user_df.loc[i, col_nam])) == next_mn:
            next_mn_fl = True
            break

    # if the next month is not available, take the last available month
    if next_mn_fl == True:
        test_mn = next_mn
    else:
        for j in range(1, 53):
            col_nam = "DataMonth" + str(j)
            if int(user_df.loc[i, col_nam]) == 0:
                break
        if j-2 >= 1:
            col_nam = "DataMonth" + str(j-2)
            test_mn = str(int(user_df.loc[i, col_nam]))
            
    user_tt_df.loc[i, "User"] = user
    user_tt_df.loc[i, "TrainMonth"] = train_mn
    user_tt_df.loc[i, "TestMonth"] = test_mn
#     if test_mn == "na":
#         list_notrain = list_notrain + user + ","
#         count_notrain = count_notrain + 1
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

pd.options.mode.chained_assignment = None
```

```python
file = "C:/Users/12sha/Documents/thesislocation/Writings/media/excels/mode_labels.xlsx"
xl = pd.ExcelFile(file)
df = xl.parse('mode_labels')
ax = df.plot.bar(x='Mode', y='Distance(KM)')
ax.legend_.remove()
plt.title('Transportation Mode Labels')
plt.ylabel('Distance(KMs)')
plt.xlabel('Mode')
plt.tight_layout()
plt.savefig(file + " - chart.png")
```
